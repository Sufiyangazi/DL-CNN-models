```python
# Translation matrix the first row moves pixels horizentally
# second row moves vertically
M = np.float32([[1,0,250],[0,2,500]])
M
```

```python
shifted = cv2.warpAffine(image,
                         M,
                         (image.shape[1],
                          image.shape[0]))
plt.imshow(shifted)
```

```python
m = np.float32([[65,43,65],
                [2,2,2]])
shift = cv2.warpAffine(image,
                       m,
                       (image.shape[1],
                        image.shape[0]))
plt.imshow(shift)
```

```python
m = np.float32([[4,5,4],
                [2,2,2]])
shift1 = cv2.warpAffine(image,
                        m,
                        (image.shape[1],
                         image.shape[0]))
plt.imshow(shift1)
```

```python
m = np.float32([[4,0,4],
                [2,0,2]])
shift1 = cv2.warpAffine(image,
                        m,
                        (image.shape[1],
                         image.shape[0]))
plt.imshow(shift1)
```

```python
m = np.float32([[1,0,-110],
                [0,1,-98]])
sf = cv2.warpAffine(image,
                    m,
                    (image.shape[1],
                     image.shape[0]))
plt.imshow(sf)
```

```python
m = np.float32([[1,43,-110],
                [0,2,-98]])
sf = cv2.warpAffine(image,
                    m,
                    (image.shape[1],
                     image.shape[0]))
plt.imshow(sf)
```

```python
(height,width,channels) = image.shape
center = (width //2,height // 2)
center
```

```python
m = cv2.getRotationMatrix2D(center,45,1.0)
rotate = cv2.warpAffine(image,m,(width,height))
plt.imshow(rotate)
```

```python
m = cv2.getRotationMatrix2D(center,90,1.0)
df = cv2.warpAffine(image,m,
                    (width,height))
plt.imshow(df)
```
